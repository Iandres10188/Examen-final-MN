```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los datos
datos = [
    (-7.0, -2.48), (-6.95, 5.21), (-6.91, 5.79), (-6.87, 7.32), (-6.83, 0.13), 
    (-6.79, 7.58), (-6.75, 3.38), (-6.71, 3.44), (-6.67, -4.62), (-6.63, -0.14),
    (-6.59, 4.26), (-6.55, 2.5), (-6.51, 3.3), (-6.47, 1.35), (-6.43, 3.87),
    (-6.39, -2.44), (-6.35, 2.17), (-6.31, 1.45), (-6.27, 0.82), (-6.23, 3.9),
    (-6.19, 2.21), (-6.15, 6.16), (-6.11, 4.64), (-6.07, 0.66), (-6.03, -1.46),
    (-5.99, 3.3), (-5.95, 0.65), (-5.91, 5.65), (-5.87, 9.45), (-5.83, -2.37),
    (-5.79, 1.8), (-5.75, 1.8), (-5.71, 6.72), (-5.67, -1.53), (-5.63, 9.06),
    (-5.59, 0.53), (-5.55, -5.18), (-5.51, 3.49), (-5.47, 7.74), (-5.43, 4.23),
    (-5.39, 2.7), (-5.35, 2.96), (-5.31, 1.15), (-5.27, 4.79), (-5.23, 2.5),
    (-5.19, 0.83), (-5.15, -6.04), (-5.11, -6.16), (-5.07, -2.36), (-5.03, -2.23),
    (-4.98, -0.9), (-4.94, 2.89), (-4.9, -2.34), (-4.86, 0.3), (-4.82, -2.89),
    (-4.78, 0.66), (-4.74, 1.0), (-4.7, 5.26), (-4.66, -0.96), (-4.62, 0.72),
    (-4.58, 6.62), (-4.54, 3.97), (-4.5, -1.09), (-4.46, 6.13), (-4.42, 4.12),
    (-4.38, 6.75), (-4.34, 9.89), (-4.3, 6.8), (-4.26, 3.55), (-4.22, 1.62),
    (-4.18, 2.22), (-4.14, 4.67), (-4.1, 3.87), (-4.06, -0.6), (-4.02, 0.07),
    (-3.98, 1.0), (-3.94, 2.38), (-3.9, 9.58), (-3.86, 5.06), (-3.82, -1.58),
    (-3.78, 4.79), (-3.74, 0.26), (-3.7, 10.75), (-3.66, 6.23), (-3.62, 4.65),
    (-3.58, 1.59), (-3.54, 6.79), (-3.5, 8.71), (-3.46, 2.85), (-3.42, 7.05),
    (-3.38, 7.83), (-3.34, 4.57), (-3.3, 5.43), (-3.26, 2.97), (-3.22, 2.61),
    (-3.18, 2.28), (-3.14, 7.98), (-3.1, 10.59), (-3.06, 9.46), (-3.02, 6.12)
]

# Extraer x e y
x = np.array([p[0] for p in datos])
y = np.array([p[1] for p in datos])

# Derivadas parciales
def derivada_polinomio_grado_n(n, grado_total):
    """Genera una función que representa la derivada parcial de un polinomio de grado n."""
    def derivada(xs, ys):
        A_fila = [sum(x**i for x in xs) for i in range(n, n + grado_total + 1)]
        b_fila = sum(y * (x**n) for x, y in zip(xs, ys))
        return A_fila + [b_fila]
    return derivada

# Definir el grado del polinomio
grado = 5

# Funciones de derivadas parciales
gradiente = [derivada_polinomio_grado_n(i, grado) for i in range(grado + 1)]

# Aplicar ajuste por mínimos cuadrados utilizando eliminación gaussiana
coeficientes = ajustar_min_cuadrados(x.tolist(), y.tolist(), gradiente)

# Definir la función polinómica con los coeficientes obtenidos
def polinomio(x_val):
    return sum(c * (x_val**i) for i, c in enumerate(coeficientes))

# Evaluar la interpolación en un rango más denso de x
x_interp = np.linspace(min(x), max(x), 200)
y_interp = [polinomio(xi) for xi in x_interp]

# Graficar los puntos originales y la interpolación
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='red', label='Datos originales', marker='o')  # Puntos originales
plt.plot(x_interp, y_interp, color='blue', linestyle='-', label='Interpolación')  # Línea interpolada
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Interpolación")
plt.legend()
plt.grid()
plt.show()
```
